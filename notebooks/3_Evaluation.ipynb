```python
import plotly.graph_objects as go
from plotly import offline
```

```python
import pandas as pd
import numpy as np
```

```python
metrics = pd.read_csv('../models/cars_classifier_metrics2.csv')
```

```python
metrics
```

```python

# Add data
epochs = metrics['epoch']
train_acc = metrics['acc']*100
train_loss = metrics['loss']
val_acc = metrics['val_acc']*100
val_loss = metrics['val_loss']

fig = go.Figure()
# Create and style traces
fig.add_trace(go.Scatter(x=epochs, y=train_acc, name='Training Accuracy',
                         line=dict(color='teal', width=4)))
fig.add_trace(go.Scatter(x=epochs, y=val_acc, name='Validation Accuracy',
                         line = dict(color='royalblue', width=4)))

# Edit the layout
fig.update_layout(title='Model Accuracy',
                   xaxis_title='Epochs',
                   yaxis_title='Accuracy (%)')

offline.plot(fig, filename = '../ptraining_accuracy.html', auto_open=False)
fig.show()
```
